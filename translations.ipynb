```python
import os
import pandas as pd
from deep_translator import GoogleTranslator
import ast
```

```python
folder1 = 'meza_dzivnieki'
folder2 = 'citi_atteli'
a1= os.listdir(folder1)
a2 = os.listdir(folder2)
a1_no_ext=[os.path.splitext(x)[0] for x in a1]
a2_no_ext=[os.path.splitext(x)[0] for x in a2]
a1_no_ext, a2_no_ext #create list of file names without extension
```

```python
to_translate1 = str(a1_no_ext)
to_translate2 = str(a2_no_ext)
translated_a1_lv = GoogleTranslator(source='auto', target='lv').translate(to_translate1)
translated_a1_lt = GoogleTranslator(source='auto', target='lt').translate(to_translate1)
translated_a2_lv = GoogleTranslator(source='auto', target='lv').translate(to_translate2)
translated_a2_lt = GoogleTranslator(source='auto', target='lt').translate(to_translate2)
translated_a1_lv, translated_a1_lt, translated_a2_lv, translated_a2_lt
```

```python
# initializing string representation of a list
ini_list = translated_a1_lv
ini_list2 = translated_a2_lv
 
# Converting string to list
lv1 = ast.literal_eval(ini_list)
lv2 = ast.literal_eval(ini_list2)
```

```python
df = pd.DataFrame(zip(a1, a1_no_ext, lv1), columns =['Image', 'English', 'Latvian'])
df
```

```python
df2 = pd.DataFrame(zip(a2, a2_no_ext, lv2), columns =['Image', 'English', 'Latvian'])
df2
```
